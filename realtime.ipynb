```python
import cv2
from deepface import DeepFace
import time

def initialize_webcam():
    cap = cv2.VideoCapture(0)  # Use 0 for default webcam
    if not cap.isOpened():
        print("Error: Could not open webcam.")
    return cap

def detect_emotions():
    cap = initialize_webcam()
    if not cap:
        return

    try:
        while True:
            ret, frame = cap.read()
            if not ret:
                print("Error: Failed to capture image.")
                break

            try:
                # Analyze the frame using DeepFace
                predictions = DeepFace.analyze(frame, actions=['emotion'])

                # Handle multiple faces detected
                for prediction in predictions:
                    emotion = prediction['dominant_emotion']
                    box = prediction['region']

                    # Extract bounding box coordinates
                    x, y, w, h = box['x'], box['y'], box['w'], box['h']

                    # Draw bounding box and display emotions on the frame
                    cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 2)
                    font = cv2.FONT_HERSHEY_SIMPLEX
                    cv2.putText(frame, emotion, (x, y - 10), font, 1, (0, 255, 0), 2, cv2.LINE_4)

            except ValueError as ve:
                print(f"ValueError: {ve}")
                cv2.putText(frame, 'No face detected', (50, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2, cv2.LINE_4)

            # Display the frame
            cv2.imshow('Real-Time Emotion Detection', frame)

            # Exit on ESC key
            if cv2.waitKey(1) & 0xFF == 27:
                break

            # Optional: Pause for a short time between frames
            time.sleep(0.1)

    finally:
        cap.release()
        cv2.destroyAllWindows()

if __name__ == "__main__":
    detect_emotions()
```
